```python
from sklearn.cluster import KMeans

coords = data[['LATITUDE','LONGITUDE']]

kmeans = KMeans(n_clusters=20, random_state=42)
data['LOCATION_CLUSTER'] = kmeans.fit_predict(coords)
```

```python
# Get latitude/longitude ranges for each cluster
cluster_ranges = data.groupby('LOCATION_CLUSTER').agg(
    LAT_MIN=('LATITUDE', 'min'),
    LAT_MAX=('LATITUDE', 'max'),
    LON_MIN=('LONGITUDE', 'min'),
    LON_MAX=('LONGITUDE', 'max')
).reset_index()

print(cluster_ranges)
```

```python
# Dealing with (0,0)-1 cluster group
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = data.dropna(subset=['LATITUDE','LONGITUDE']).copy()
data = data[~((data['LATITUDE'] == 0) & (data['LONGITUDE'] == 0))].copy()

print(f"Remaining rows after cleaning: {data.shape[0]}")

coords = data[['LATITUDE', 'LONGITUDE']]

kmeans = KMeans(n_clusters=20, random_state=42)
data['LOCATION_CLUSTER'] = kmeans.fit_predict(coords)
```

```python


cluster_ranges = data.groupby('LOCATION_CLUSTER').agg(
    LAT_MIN=('LATITUDE', 'min'),
    LAT_MAX=('LATITUDE', 'max'),
    LON_MIN=('LONGITUDE', 'min'),
    LON_MAX=('LONGITUDE', 'max')
).reset_index()

cluster_centers = data.groupby('LOCATION_CLUSTER')[['LATITUDE','LONGITUDE']].mean().reset_index()
cluster_info = cluster_ranges.merge(cluster_centers, on='LOCATION_CLUSTER')
cluster_info.rename(columns={'LATITUDE':'LAT_CENTER','LONGITUDE':'LON_CENTER'}, inplace=True)

print(cluster_info)

```

```python
#  Scatter plot to visualize clusters
plt.figure(figsize=(10,8))
scatter = plt.scatter(
    data['LONGITUDE'],
    data['LATITUDE'],
    c=data['LOCATION_CLUSTER'],
    s=5,
    alpha=0.6,
    cmap='tab20'
)
plt.colorbar(scatter, label='Cluster ID')
plt.title("Chicago Crash Data Spatial Clusters (KMeans)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xlim(-88, -87.5)
plt.ylim(41.6, 42.1)
plt.show()
```

Applying Label Encoding to the specified categorical features and then dropping the columns identified for removal. Missing values in categorical columns will be filled with 'Missing' before encoding to prevent errors.

```python
from sklearn.preprocessing import LabelEncoder

data_labeled = data.copy()

# Apply Label Encoding to specified categorical columns
for col in cat_columns:
    if col in data_labeled.columns:
        # Fill NaN values before encoding, as LabelEncoder does not handle them
        data_labeled[col] = data_labeled[col].fillna('Missing')
        le = LabelEncoder()
        data_labeled[col] = le.fit_transform(data_labeled[col])
        print(f"Column '{col}' encoded.")
```

```python
# Dealing with temporal data
import pandas as pd
import numpy as np
from datetime import datetime, date # Import datetime and date modules


# Time of day (categorical)
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

data_labeled['CRASH_TIME_OF_DAY'] = data_labeled['CRASH_HOUR'].apply(time_of_day)
```
